```python
import matplotlib.pyplot as mp
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from pylab import *
```

```python
x=np.linspace(0,2*np.pi)
y=np.sin(x)
y2=np.cos(x)
mp.xlabel("x")
mp.ylabel("sin(x)")
mp.title("plot of the sine function")
mp.plot(x,y, label="sin")
mp.plot(x,y2,"r--", label="cos")
mp.legend()
```

```python
x=np.outer(np.linspace(-2,2,20),np.ones(20))
y=x.copy().T
x,y= np.meshgrid(x,y)
z=x*np.exp(-x**2-y**2)
fig,ax= mp.subplots(subplot_kw={"projection":"3d"})
surf = ax.plot_surface(x,y,z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
figure,axis=mp.subplots(2,3, figsize=(9,6.5))
x=np.linspace(1e-50,2*np.pi)
xlim(x.min()*1.1,x.max()*1.1)
ylim(-4,4)

y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)

axis[0,0].plot(x,y1)
axis[0,0].set_title("Sine")
axis[0,1].set_xlim([x.min()*1.1,x.max()*1.1])
axis[0,1].set_ylim([-1,1])

axis[0,1].plot(x,y2)
axis[0,1].set_title("Cosine")
axis[0,1].set_xlim([x.min()*1.1,x.max()*1.1])
axis[0,1].set_ylim([-1,1])

axis[0,2].plot(x,y3)
axis[0,2].set_title("Tangent")
axis[0,2].set_xlim([x.min()*1.1,x.max()*1.1])
axis[0,2].set_ylim([-4,4])

axis[1,0].plot(x,1/y1)
axis[1,0].set_title("Cosecant")
axis[1,0].set_xlim([x.min()*1.1,x.max()*1.1])
axis[1,0].set_ylim([-4,4])

axis[1,1].plot(x,1/y2)
axis[1,1].set_title("Secant")
axis[1,1].set_xlim([x.min()*1.1,x.max()*1.1])
axis[1,1].set_ylim([-4,4])

axis[1,2].plot(x,1/y3)
axis[1,2].set_title('Cotangent')
axis[1,2].set_xlim([x.min()*1.1,x.max()*1.1])
axis[1,2].set_ylim([-4,4])
```
